```python
from scipy.io.wavfile import read #import the required function from the module
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
import numpy as np
```

```python
samplerate1, data1 = read('jnt_stan.wav')
samplerate2, data2 = read('dhl_stan.wav')
samplerate3, data3 = read('jnt_wee.wav')
samplerate4, data4 = read('dhl_wee.wav')
samplerate5, data5 = read('jnt_khai.wav')
samplerate6, data6 = read('dhl_khai.wav')
```

```python
sampleratelong, datalong = read('sample_wee.wav')
```

```python
newdata = np.array_split(datalong, 8)
    
```

```python
for data in newdata:
    distance, path = fastdtw(data3, data, dist=euclidean)
    print(distance)
```

```python
import matplotlib.pyplot as plt
import numpy as np

for data in newdata:
    duration = len(data)/samplerate3
    time = np.arange(0,duration,1/samplerate3) #time vector

    plt.plot(time,data)
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.title('sample_wee.wav')
    plt.show()
```
